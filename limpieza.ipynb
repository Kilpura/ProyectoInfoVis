```python
import pandas as pd
import os

def guardar_en_csv(df_filtrado, nombre_archivo):
    '''
    Args:
        df_filtrado: DataFrame con las columnas filtradas
        nombre_archivo (str): nombre del archivo de salida en csv
    '''

    df_filtrado.to_csv(nombre_archivo, index=False)

def abrir_csv(path):
    
    df = pd.read_csv(path)
    return df

path_archivo = os.path.join('data sucias','Free_Press_Index_2019_2024.csv')
df = abrir_csv(path_archivo)

df.info()
```

```python
df1 = pd.read_csv(os.path.join('data sucias', 'countries_gdp_hist.csv'), sep=';')

df1
```

```python
df1.info()
df1 = df1.drop('region_name', axis=1)
df1 = df1.drop('sub_region_name', axis=1)
df1 = df1.drop('intermediate_region', axis=1)
df1
```

```python
df1 = df1[df1['year'].isin([2019, 2020, 2021, 2022, 2023])]
df1['country_name'] = df1['country_name'].str.title()
df1['income_group'] = df1['income_group'].str.title()

df1.rename(columns={'country_name': 'Country'}, inplace=True)

df1
```

```python
df_2019 = df1[df1['year'] == 2019]
df_2020 = df1[df1['year'] == 2020]
df_2021 = df1[df1['year'] == 2021]
df_2022 = df1[df1['year'] == 2022]
df_2023 = df1[df1['year'] == 2023]
```

```python
df_2019.rename(columns={
    'country_code' : '2019_country_code',
    'income_group': '2019_income_group',
    'total_gdp': '2019_total_gdp',
    'total_gdp_million': '2019_total_gdp_million',
    'gdp_variation': '2019_gdp_variation'
}, inplace=True)

df_2019.drop('year', axis=1, inplace=True)
```

```python
df_2020.rename(columns={
    'country_code' : '2020_country_code',
    'income_group': '2020_income_group',
    'total_gdp': '2020_total_gdp',
    'total_gdp_million': '2020_total_gdp_million',
    'gdp_variation': '2020_gdp_variation'
}, inplace=True)

df_2020.drop('year', axis=1, inplace=True)
```

```python
df_2021.rename(columns={
    'country_code' : '2021_country_code',
    'income_group': '2021_income_group',
    'total_gdp': '2021_total_gdp',
    'total_gdp_million': '2021_total_gdp_million',
    'gdp_variation': '2021_gdp_variation'
}, inplace=True)

df_2021.drop('year', axis=1, inplace=True)
```

```python
df_2022.rename(columns={
    'country_code' : '2022_country_code',
    'income_group': '2022_income_group',
    'total_gdp': '2022_total_gdp',
    'total_gdp_million': '2022_total_gdp_million',
    'gdp_variation': '2022_gdp_variation'
}, inplace=True)

df_2022.drop('year', axis=1, inplace=True)
```

```python
df_2023.rename(columns={
    'country_code' : '2023_country_code',
    'income_group': '2023_income_group',
    'total_gdp': '2023_total_gdp',
    'total_gdp_million': '2023_total_gdp_million',
    'gdp_variation': '2023_gdp_variation'
}, inplace=True)

df_2023.drop('year', axis=1, inplace=True)
```

```python
df_2019 = df_2019[['2019_country_code', 'Country', '2019_income_group', '2019_total_gdp', '2019_total_gdp_million', '2019_gdp_variation']]
df_2020 = df_2020[['2020_country_code','Country', '2020_income_group', '2020_total_gdp', '2020_total_gdp_million', '2020_gdp_variation']]
df_2021 = df_2021[['2021_country_code','Country', '2021_income_group', '2021_total_gdp', '2021_total_gdp_million', '2021_gdp_variation']]
df_2022 = df_2022[['2022_country_code','Country', '2022_income_group', '2022_total_gdp', '2022_total_gdp_million', '2022_gdp_variation']]
df_2023 = df_2023[['2023_country_code','Country', '2023_income_group', '2023_total_gdp', '2023_total_gdp_million', '2023_gdp_variation']]

# Concatenamos los DataFrames por filas, sin duplicar columnas
df_pib_tiempo = pd.concat([df_2019.set_index('Country'), 
                           df_2020.set_index('Country'), 
                           df_2021.set_index('Country'), 
                           df_2022.set_index('Country'), 
                           df_2023.set_index('Country')], 
                          axis=1, join='inner')

# Reseteamos el índice (para mantener la columna 'Country')
df_pib_tiempo.reset_index(inplace=True)
df_pib_tiempo
```

```python
guardar_en_csv(df_pib_tiempo, 'data limpia/Countries_GDP_CLEEN.csv')
```
