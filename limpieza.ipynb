```python
import pandas as pd
import os

def abrir_csv(path):
    
    df = pd.read_csv(path)
    return df

path_archivo = os.path.join('Free_Press_Index_2019_2024.csv')
df = abrir_csv(path_archivo)

df.info()
```

```python
df1 = pd.read_csv('countries_gdp_hist.csv', sep=';')

df1
```

```python
df1.info()
df1 = df1.drop('country_code', axis=1)
df1 = df1.drop('region_name', axis=1)
df1 = df1.drop('sub_region_name', axis=1)
df1 = df1.drop('intermediate_region', axis=1)
```

```python
df1 = df1[df1['year'].isin([2019, 2020, 2021, 2022, 2023])]
df1['country_name'] = df1['country_name'].str.title()
df1['income_group'] = df1['income_group'].str.title()
df1.rename(columns={'country_name': 'Country'}, inplace=True)

df1
```

```python
df_2019 = df1[df1['year'] == 2019]
df_2020 = df1[df1['year'] == 2020]
df_2021 = df1[df1['year'] == 2021]
df_2022 = df1[df1['year'] == 2022]
df_2023 = df1[df1['year'] == 2023]
```

```python
df_2019.rename(columns={
    'income_group': '2019_income_group',
    'total_gdp': '2019_total_gdp',
    'total_gdp_million': '2019_total_gdp_million',
    'gdp_variation': '2019_gdp_variation'
}, inplace=True)

df_2019.drop('year', axis=1, inplace=True)
```

```python
df_2020.rename(columns={
    'income_group': '2020_income_group',
    'total_gdp': '2020_total_gdp',
    'total_gdp_million': '2020_total_gdp_million',
    'gdp_variation': '2020_gdp_variation'
}, inplace=True)

df_2020.drop('year', axis=1, inplace=True)
```

```python
df_2021.rename(columns={
    'income_group': '2021_income_group',
    'total_gdp': '2021_total_gdp',
    'total_gdp_million': '2021_total_gdp_million',
    'gdp_variation': '2021_gdp_variation'
}, inplace=True)

df_2021.drop('year', axis=1, inplace=True)
```

```python
df_2022.rename(columns={
    'income_group': '2022_income_group',
    'total_gdp': '2022_total_gdp',
    'total_gdp_million': '2022_total_gdp_million',
    'gdp_variation': '2022_gdp_variation'
}, inplace=True)

df_2022.drop('year', axis=1, inplace=True)
```

```python
df_2023.rename(columns={
    'income_group': '2023_income_group',
    'total_gdp': '2023_total_gdp',
    'total_gdp_million': '2023_total_gdp_million',
    'gdp_variation': '2023_gdp_variation'
}, inplace=True)

df_2023.drop('year', axis=1, inplace=True)
```

```python
df_2019_merged = df.merge(df_2019, on='Country', how='left')
df_2019_merged
```

```python
cols_2019 = ['2019_income_group', '2019_total_gdp', '2019_total_gdp_million', '2019_gdp_variation']
pos = df_2019_merged.columns.get_loc('2019_Score') + 1
cols = list(df_2019_merged.columns)
for col in cols_2019:
    cols.remove(col)
for i, col in enumerate(cols_2019):
    cols.insert(pos + i, col)
df_2019_merged = df_2019_merged[cols]
df_2019_merged
```

```python
df_2020_merged = df_2019_merged.merge(df_2020, on='Country', how='left')
cols_2020 = ['2020_income_group', '2020_total_gdp', '2020_total_gdp_million', '2020_gdp_variation']
pos = df_2020_merged.columns.get_loc('2020_Score') + 1
cols = list(df_2020_merged.columns)
for col in cols_2020:
    cols.remove(col)
for i, col in enumerate(cols_2020):
    cols.insert(pos + i, col)
df_2020_merged = df_2020_merged[cols]
df_2020_merged
```

```python
df_2021_merged = df_2020_merged.merge(df_2021, on='Country', how='left')
cols_2021 = ['2021_income_group', '2021_total_gdp', '2021_total_gdp_million', '2021_gdp_variation']
pos = df_2021_merged.columns.get_loc('2021_Score') + 1
cols = list(df_2021_merged.columns)
for col in cols_2021:
    cols.remove(col)
for i, col in enumerate(cols_2021):
    cols.insert(pos + i, col)
df_2021_merged = df_2021_merged[cols]
df_2021_merged
```

```python
df_2022_merged = df_2021_merged.merge(df_2022, on='Country', how='left')
cols_2022 = ['2022_income_group', '2022_total_gdp', '2022_total_gdp_million', '2022_gdp_variation']
pos = df_2022_merged.columns.get_loc('2022_Score') + 1
cols = list(df_2022_merged.columns)
for col in cols_2022:
    cols.remove(col)
for i, col in enumerate(cols_2022):
    cols.insert(pos + i, col)
df_2022_merged = df_2022_merged[cols]
df_2022_merged
```

```python
df_2023_merged = df_2022_merged.merge(df_2023, on='Country', how='left')
cols_2023 = ['2023_income_group', '2023_total_gdp', '2023_total_gdp_million', '2023_gdp_variation']
pos = df_2023_merged.columns.get_loc('2023_Score') + 1
cols = list(df_2023_merged.columns)
for col in cols_2023:
    cols.remove(col)
for i, col in enumerate(cols_2023):
    cols.insert(pos + i, col)
df_2023_merged = df_2023_merged[cols]
df_2023_merged
```

```python
def guardar_en_csv(df_filtrado, nombre_archivo):
    '''
    Args:
        df_filtrado: DataFrame con las columnas filtradas
        nombre_archivo (str): nombre del archivo de salida en csv
    '''

    df_filtrado.to_csv(nombre_archivo, index=False)

guardar_en_csv(df_2023_merged, 'Prensa_Ingreso.csv')
```
