```python
df1 = pd.read_csv(os.path.join('data sucias', 'countries_gdp_hist.csv'), sep=';')

df1
```

```python
df1.info()
df1 = df1.drop('region_name', axis=1)
df1 = df1.drop('sub_region_name', axis=1)
df1 = df1.drop('intermediate_region', axis=1)
df1 = df1.drop('total_gdp', axis=1)
df1 = df1.drop('gdp_variation', axis=1)
df1
```

```python
df1 = df1[df1['year'].isin([2019, 2020, 2021, 2022, 2023])]
df1['country_name'] = df1['country_name'].str.title()
df1['income_group'] = df1['income_group'].str.title()

df1.rename(columns={'country_name': 'Country'}, inplace=True)

df1
```

```python
df_2019 = df1[df1['year'] == 2019]
df_2020 = df1[df1['year'] == 2020]
df_2021 = df1[df1['year'] == 2021]
df_2022 = df1[df1['year'] == 2022]
df_2023 = df1[df1['year'] == 2023]

def agregar_columna_porcentaje(df, año):
    col_gdp = f"{año}_total_gdp_million"
    col_pct = f"{año}_porcentaje"
    
    df[col_gdp] = pd.to_numeric(df[col_gdp], errors='coerce')
    
    min_val = df[col_gdp].min()
    max_val = df[col_gdp].max()
    
    df[col_pct] = ((df[col_gdp] - min_val) / (max_val - min_val)) * 100
    df[col_pct] = df[col_pct].round(2)
    
    # Reubicar la  justo después de la original
    cols = df.columns.tolist()
    idx = cols.index(col_gdp)
    cols.insert(idx + 1, cols.pop(cols.index(col_pct)))
    df = df[cols]
    
    return df
```

```python
df_2019.rename(columns={
    'country_code' : '2019_country_code',
    'income_group': '2019_income_group',
    'total_gdp_million': '2019_total_gdp_million',
}, inplace=True)

df_2019.drop('year', axis=1, inplace=True)
```

```python
df_2019_porcentaje = agregar_columna_porcentaje(df_2019, 2019)
```

```python
df_2021.rename(columns={
    'country_code' : '2021_country_code',
    'income_group': '2021_income_group',
    'total_gdp_million': '2021_total_gdp_million',
}, inplace=True)

df_2021.drop('year', axis=1, inplace=True)

df_2021_porcentaje = agregar_columna_porcentaje(df_2021, 2021)
```
