```python
import pandas as pd
import os
from fuzzywuzzy import process
import numpy as np



def abrir_csv(path):
    
    df = pd.read_csv(path, sep=';')
    return df

def abrir_csv_1(path):

    df = pd.read_csv(path, sep=',')
    return df

def guardar_excel(path, df):
    df.to_excel(path, index=False)

def guardar_csv(path, df):
    df.to_csv(path, sep=';', index=False)
```

```python
path_archivo_countries = os.path.join('archivos_sucios', 'countries_gdp_hist.csv')
path_archivo_press = os.path.join('archivos_sucios', 'Free_Press_Index_2019_2024.csv')

df1 = abrir_csv(path_archivo_countries)
df2 = abrir_csv_1(path_archivo_press)
```

```python
df1
```

```python
# botar columnas
df1.drop(columns=['region_name', 'sub_region_name', 'intermediate_region', 'total_gdp_million'], inplace=True)

# filtrar x años q necesitamos
df1 = df1[df1['year'].isin([2019, 2020, 2021, 2022, 2023, 2024])]

# renombrar columna
df1.rename(columns={'country_name': 'Country'}, inplace=True)


df1
```

```python
# renombrar la primera columna de df2 a 'ranking'
df2.rename(columns={'Unnamed: 0': 'ranking'}, inplace=True)

df2
```

```python
df_2019 = df1[df1['year'] == 2019]
df_2020 = df1[df1['year'] == 2020]
df_2021 = df1[df1['year'] == 2021]
df_2022 = df1[df1['year'] == 2022]
df_2023 = df1[df1['year'] == 2023]
df_2024 = df1[df1['year'] == 2024]

df_2019.rename(columns={
    'income_group': '2019_income_group',
    'total_gdp': '2019_total_gdp',
}, inplace=True)

df_2019.drop('year', axis=1, inplace=True)

df_2020.rename(columns={
    'income_group': '2020_income_group',
    'total_gdp': '2020_total_gdp',
}, inplace=True)

df_2020.drop('year', axis=1, inplace=True)

df_2021.rename(columns={
    'income_group': '2021_income_group',
    'total_gdp': '2021_total_gdp',
}, inplace=True)

df_2021.drop('year', axis=1, inplace=True)

df_2022.rename(columns={
    'income_group': '2022_income_group',
    'total_gdp': '2022_total_gdp',
}, inplace=True)

df_2022.drop('year', axis=1, inplace=True)

df_2023.rename(columns={
    'income_group': '2023_income_group',
    'total_gdp': '2023_total_gdp',
}, inplace=True)

df_2023.drop('year', axis=1, inplace=True)

df_2024.rename(columns={
    'income_group': '2024_income_group',
    'total_gdp': '2024_total_gdp',
}, inplace=True)

df_2024.drop('year', axis=1, inplace=True)
```

```python
df_2019 = df_2019[['country_code', 'Country', '2019_income_group', '2019_total_gdp']]
df_2020 = df_2020[['country_code', '2020_income_group', '2020_total_gdp']]
df_2021 = df_2021[['country_code', '2021_income_group', '2021_total_gdp']]
df_2022 = df_2022[['country_code', '2022_income_group', '2022_total_gdp']]
df_2023 = df_2023[['country_code', '2023_income_group', '2023_total_gdp']]
df_2024 = df_2024[['country_code', '2024_income_group', '2024_total_gdp']]

df_countries = pd.concat([
    df_2019.set_index('country_code'),
    df_2020.set_index('country_code'),
    df_2021.set_index('country_code'),
    df_2022.set_index('country_code'),
    df_2023.set_index('country_code')
], axis=1, join='inner').reset_index()  # Restablecer el índice para que 'country_code' sea una columna

df_countries
```

```python
# quiero que todos los country esten en minuscula con la primera en mayuscula
df_countries['Country'] = df_countries['Country'].str.lower().str.capitalize()
float_cols = df_countries.select_dtypes(include='float64').columns

for col in float_cols:
    df_countries[col] = df_countries[col].round(0).astype('Int64')

df_countries['Country'] = df_countries['Country'].astype(str)
df_countries['country_code'] = df_countries['country_code'].astype(str)
df_countries
```

```python
countries_list = df_countries['Country'].unique()
press_countries = df2['Country'].unique()

country_name_mapping = {}
for name in press_countries:
    match, score = process.extractOne(name, countries_list)
    if score >= 90:
        country_name_mapping[name] = match

df2['Country'] = df2['Country'].replace(country_name_mapping)

float_cols = df2.select_dtypes(include='float64').columns

for col in float_cols:
    df2[col] = df2[col].round(0).astype('Int64')

df2['Country'] = df2['Country'].astype(str)

df2['ranking'] = df2['ranking'].astype('Int64')

df2
```

```python
# quiero hacer merge de ambos df en base al country 
df_merged = pd.merge(df_countries, df2, on='Country', how='inner')

df_merged = df_merged[['ranking', 'Country', 'country_code', '2019_Rank', '2019_Score', '2019_income_group', '2019_total_gdp', '2020_Rank', '2020_Score', '2020_income_group', '2020_total_gdp', '2021_Rank', '2021_Score', '2021_income_group', '2021_total_gdp', '2022_Rank', '2022_Score', '2022_income_group', '2022_total_gdp', '2023_Rank', '2023_Score', '2023_income_group', '2023_total_gdp', '2024_Rank', '2024_Score']]

guardar_excel(os.path.join('archivos_merge', 'merged_NaN.xlsx'), df_merged)
```

```python
valores_nulos = [0, "0", '', ' ', 'nan', 'NaN', '-', 'N/A', 'n/a', 'NULL', 'null', None, 'None', 'none']


df_merged = df_merged.replace(valores_nulos, np.nan)

df_merged_sin_NaN = df_merged.dropna(
    subset=['2019_income_group', '2020_income_group', '2021_income_group', '2022_income_group', '2023_income_group'],
    how='all'
)

# Luego eliminar filas sin total_gdp en todos los años, usando el resultado anterior
df_merged_sin_NaN = df_merged_sin_NaN.dropna(
    subset=['2019_total_gdp', '2020_total_gdp', '2021_total_gdp', '2022_total_gdp', '2023_total_gdp'],
    how='all'
)

# Reemplazar NaN en las columnas 'total_gdp' con un valor predeterminado, como 0
df_merged_sin_NaN[['2019_total_gdp', '2020_total_gdp', '2021_total_gdp', '2022_total_gdp', '2023_total_gdp']] = df_merged_sin_NaN[['2019_total_gdp', '2020_total_gdp', '2021_total_gdp', '2022_total_gdp', '2023_total_gdp']].fillna(0)
df_merged_sin_NaN = df_merged_sin_NaN.replace(valores_nulos, np.nan)

df_merged_sin_NaN['ranking'] = df_merged_sin_NaN['ranking'].fillna(0)

df_merged_sin_NaN
```

```python
df_merged_sin_NaN.isnull().sum()
```

```python
df_merged_sin_NaN[['2019_total_gdp', '2020_total_gdp', '2021_total_gdp', '2022_total_gdp', '2023_total_gdp']] = df_merged_sin_NaN[['2019_total_gdp', '2020_total_gdp', '2021_total_gdp', '2022_total_gdp', '2023_total_gdp']].fillna(0)
df_merged_sin_NaN = df_merged_sin_NaN.dropna(how='all')
df_merged_sin_NaN = df_merged_sin_NaN[
    ~((df_merged_sin_NaN[['2019_total_gdp', '2020_total_gdp', '2021_total_gdp', '2022_total_gdp', '2023_total_gdp']] == 0).all(axis=1))
]

df_merged_sin_NaN
```

```python
guardar_excel(os.path.join('archivos_merge', 'merged_sin_NaN.xlsx'), df_merged_sin_NaN)
guardar_csv(os.path.join('archivos_merge', 'merged_sin_NaN.csv'), df_merged_sin_NaN)
```
